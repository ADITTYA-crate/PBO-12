```python
# =========================================
# IMPORT LIBRARY
# =========================================

# dataclass untuk membuat class lebih sederhana
from dataclasses import dataclass

# ABC digunakan untuk abstraction/interface
from abc import ABC, abstractmethod

# NumPy untuk operasi matematika
import numpy as np


# =========================================
# ABSTRACT CLASS / INTERFACE
# =========================================

class BangunRuang(ABC):

    # method abstract
    # wajib dimiliki semua turunan class
    @abstractmethod
    def volume(self) -> float:
        pass

    # method biasa
    def info(self):
        print("Program menghitung volume bangun ruang")


# =========================================
# CLASS KUBUS
# =========================================

@dataclass
class Kubus(BangunRuang):

    # attribute
    sisi: float

    # encapsulation
    # private method
    def __cek_input(self):
        if self.sisi <= 0:
            raise ValueError("Sisi harus lebih dari 0")

    # overriding method
    def volume(self) -> float:

        # validasi input
        self.__cek_input()

        # rumus volume kubus
        hasil = np.power(self.sisi, 3)

        return hasil


# =========================================
# CLASS BALOK
# =========================================

@dataclass
class Balok(BangunRuang):

    panjang: float
    lebar: float
    tinggi: float

    def __cek_input(self):

        if self.panjang <= 0:
            raise ValueError("Panjang harus > 0")

        if self.lebar <= 0:
            raise ValueError("Lebar harus > 0")

        if self.tinggi <= 0:
            raise ValueError("Tinggi harus > 0")

    def volume(self) -> float:

        self.__cek_input()

        hasil = self.panjang * self.lebar * self.tinggi

        return hasil


# =========================================
# CLASS BOLA
# =========================================

@dataclass
class Bola(BangunRuang):

    jari_jari: float

    def __cek_input(self):

        if self.jari_jari <= 0:
            raise ValueError("Jari-jari harus > 0")

    def volume(self) -> float:

        self.__cek_input()

        hasil = (4/3) * np.pi * np.power(self.jari_jari, 3)

        return hasil


# =========================================
# CLASS TABUNG
# =========================================

@dataclass
class Tabung(BangunRuang):

    jari_jari: float
    tinggi: float

    def __cek_input(self):

        if self.jari_jari <= 0:
            raise ValueError("Jari-jari harus > 0")

        if self.tinggi <= 0:
            raise ValueError("Tinggi harus > 0")

    def volume(self) -> float:

        self.__cek_input()

        hasil = np.pi * np.power(self.jari_jari, 2) * self.tinggi

        return hasil


# =========================================
# CLASS KERUCUT
# =========================================

@dataclass
class Kerucut(BangunRuang):

    jari_jari: float
    tinggi: float

    def __cek_input(self):

        if self.jari_jari <= 0:
            raise ValueError("Jari-jari harus > 0")

        if self.tinggi <= 0:
            raise ValueError("Tinggi harus > 0")

    def volume(self) -> float:

        self.__cek_input()

        hasil = (1/3) * np.pi * np.power(self.jari_jari, 2) * self.tinggi

        return hasil


# =========================================
# MEMBUAT OBJECT
# =========================================

kubus = Kubus(5)

balok = Balok(10, 5, 4)

bola = Bola(7)

tabung = Tabung(7, 10)

kerucut = Kerucut(7, 10)


# =========================================
# POLYMORPHISM
# =========================================

# semua object dimasukkan ke list
data_bangun = [
    kubus,
    balok,
    bola,
    tabung,
    kerucut
]


# =========================================
# MENAMPILKAN HASIL
# =========================================

for bangun in data_bangun:

    # nama class
    nama = bangun.__class__.__name__

    print(f"\nBangun Ruang : {nama}")

    print(f"Volume : {bangun.volume():.2f}")
```
